```julia
using LinearAlgebra

function GSO(B::Matrix{Int64}, N::Int64)::Tuple{Matrix{Float64},Matrix{Float64}}
    B_p = zeros(Float64,size(B))
    mu = zeros(Float64,(N,N))
    for i in 1:N
        B_p[i,:] = B[i,:]
        mu[i,i] = 1
        for j in 1:i-1
            mu[i,j] = B[i,:]â‹…B_p[j,:]/norm(B_p[j,:])^2
            B_p[i,:] -= mu[i,j]*B_p[j,:]
        end
    end
    return B_p,mu
end


function ENUM(B::Matrix{Int64},R::Float64,g::Int64,h::Int64)
    N,_ = size(B)
    Bp,mu = GSO(B,N)
    norm_Bp = zeros(Float64,N)
    for i in 1:N
        norm_Bp[i] = norm(Bp[i])^2
    end
    Bp,mu = GSO(B,N)
    sigma = zeros(Float64,(N+1,N))
    r = [i for i in 1:N]
    r0 = 0
    rho = zeros(Float64,N+1)
    v = zeros(Int64,N)
    v[g] = 1
    c = zeros(Float64,N)
    w = zeros(Int64,N)
    last_nonzero = 1
    k = g+1
    flag = 0
    v1 = zeros(Int64,N)
    
    while true
        rho[k] = rho[k+1] + ((v[k]-c[k])^2)*norm_Bp[k]
        if rho[k] <= R
            if k == g+1
                R = 0.99*rho[k]
                flag +=1
                for i in 1:N
                    v1[i] = v[i]
                end
            end
            k -=1
            ##r
            if k== 1
                r0 = max(r0,r[k])
            else
                r[k-1] = max(r[k-1],r[k])
            end
            ##
            for i in r[k]:-1:k+1
                sigma[i,k] = sigma[i+1,k]+v[i]*mu[i,k]
            end
            c[k] = -sigma[k+1,k]
            v[k] = round(c[k])
            w[k] = 1
        else
            k = k+1
            if k == h+1
                if flag == 0
                    return false
                else
                    vv = v1[g]*B[g,:]
                    for i in g+1:h
                        vv += v1[i]*B[i,:]
                    end
                    return vv
                end
            end
            ##r
            if k==1
                r0 = 1
            else
                r[k-1] = k
            end
            ##
            if k >= last_nonzero
                last_nonzero = k
                v[k] = v[k]+1
            else
                if v[k] > c[k]
                    v[k] = v[k] - w[k]
                else
                    v[k] = v[k] + w[k]
                end
                w[k] = w[k] + 1
            end
        end
    end
end

            
            
```
