```python
def MLLL(B,delta):
    h = B.nrows()
    N = B.ncols()
    z = h-1
    g =0
    B_prime = Matrix(QQ,h,N)
    norm_Bp = vector(QQ,h)
    mu = Matrix(QQ,h)
    while g <= z:
        if B[g].norm() == 0:
            if g<z:
                B[g],B[z] = B[z], B[g]
            z -=1
        B_prime[g] = B[g]
        #g行GS
        for j in range(g-1):
            mu[g,j] = B[g].inner_product(B_prime[j])/B_prime[j].norm()^2
            B_prime[g] -=mu[g,j]*B_prime[j]
        norm_Bp[g] = B_prime[g].norm()^2
        mu[g,g] = 1
        
        if g == 0:
            g = 1
        else:
            print(g)
            l = g; k=g; startgain = False
            while k <= l and (not startgain) :
                
                #Size_reduce
                if abs(mu[k,k-1]) >0.50:
                    q = round(mu[k,k-1])
                    B[k] -= q*B[k-1]
                    for l in range(k):####?
                        mu[k,l] -=q*mu[k-1,l]
                
                nu=mu[k,k-1]
                beta = norm_Bp[k] + (nu^2)*norm_Bp[k-1]
                if beta >= delta*norm_Bp[k-1]:
                    for j in reversed(range(k-2)):
                        #Size_reduce
                        if abs(mu[k,j]) >0.50:
                            q = round(mu[k,j])
                            B[k] -= q*B[j]
                            for l in range(j+1):####?
                                mu[k,l] -=q*mu[j,l]
                    k+=1
                else:
                    print(B[k])
                    if B[k].norm() == 0:
                        if k<z:
                            B[z],B[k] = B[k],B[z]
                        z-=1;g=k;startgain=True
                    else:
                        B[k-1],B[k] = B[k],B[k-1]
                        for j in range(k-2):
                            mu[k,j],mu[k-1,j] = mu[k-1,j],mu[k,j]
                        if beta!= 0:
                            if B[k] == 0:
                                B[k-1] = beta
                                B_prime[k-1] = nu*B_prime[k-1]
                                mu[k,k-1] = 1/nu
                                for i in range(k+1,l+1):
                                    mu[i,k-1] = mu[i,k-1]/nu
                            else:
                                t = norm_Bp[k-1]/beta
                                mu[k,k-1] = nu*t
                                w = B_prime[k-1]
                                B_prime[k-1] = B_prime[k] + nu*w
                                B[k-1] = beta
                                if k <= l:
                                    B_prime[k] = -mu[k,k-1]*B_prime[k] + (norm_Bp[k]/beta)*w
                                    B[k] *= t
                                
                                for i in range(k+1,l+1):
                                    t = mu[i,k]
                                    mu[i,k] = mu[i,k-1] - nu*t
                                    mu[i,k-1] = t + mu[k,k-1]*mu[i,k]
                 
                                
                        else:#beta==0
                            B[k],B[k-1] = B[k-1],B[k]
                            B_prime[k],B_prime[k-1] = B_prime[k-1],B_prime[k]
                            for i in range(k+1,l+1):
                                mu[i,k],mu[i,k-1] = mu[i,k-1],mu[i,k]          
                    k=max(k-1,1)
        
            if not startgain :
                g +=1
    return B

            
BB = Matrix(QQ,[[388,417,417,-86],[-672,-73,-121,944],[-689,379,724,653],[-179,96,-24,978],[508,-705,173,-343]])
print(MLLL(BB,0.75))
                          
```
